```python
#Since pvalue for the groups are not all >0.05,CHECK FOR NORMALITY OF RESIDUALS
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro
#Fit a linear regression model
x=sm.add_constant(np.arange(len(data_internal_temp)))
model=sm.OLS(data_internal_temp['Temperature'],x).fit()

#Get the residuals
residuals=model.resid

#visualize the residual for each group
plt.figure(figsize=(12,6))

#Boxplot
sns.boxplot(x='Time',y=residuals,data=data_internal_temp)
plt.title('Box of residuals for each group')

#swarmplot for individual data points
sns.swarmplot(x='Time',y=residuals,data=data_internal_temp,color='.25')
plt.title('Swarmplot of residuals for each group')
plt.savefig(r"C:\Users\FAITH\Documents\work\AI\klexy analysis\residual normality check for int Temp.png")
plt.show()

#shapiro_wick test for normality for each group
shapiro_results=data_internal_temp.groupby('Time')['Temperature'].apply(lambda x:stats.shapiro(x)[1])
shapiro_results=shapiro_results.reset_index()
shapiro_results.columns=['Time','p-value']

print('shapiro_wick test for normality(p-values):')
print(shapiro_results)
```

```python
#if normality of residuals is <0.05, run posthoc analysis
from scipy .stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
#perform tukey test
tukey=pairwise_tukeyhsd(endog=data_internal_temp['Temperature'],groups=data_internal_temp['Time'],alpha=0.05)
#display result
print(tukey)
```

```python
formula= 'Temperature~ Time'
model= sm.OLS.from_formula(formula, data=data_internal_temp).fit()

#Display summary
print(model.summary())
print('end of summary')

#ANOVA
anova_table= sm.stats.anova_lm(model, typ=1)
print(anova_table)
```
